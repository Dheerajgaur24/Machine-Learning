```python
"""
Linear regression
1- finds a pattern in old data
2- straight Line
3- Line
y=m*x+b

1- its not about the line, just for story
2- 5 rows use start
3- accuracy is not always the goal
"""

from sklearn.linear_model import LinearRegression

X=[[1],[2],[3],[4],[5]]
y=[40, 50, 65, 75, 90]

model=LinearRegression()

model.fit(X, y)

hours = float(input("Enter how many hours you studies ="))

predicted_marks = model.predict([[hours]])

print(f"Based on your marks {hours} you may score around {predicted_marks}")
```
